```python
import matplotlib.pyplot as plt

# Initialize empty lists for words and frequencies
words = []
frequencies = []

# Read the file containing top 50 words and their frequencies
with open('top_values', 'r') as file:
    for line in file:
        # Split the line into word and frequency
        word, freq = line.split()
        words.append(word)             # Append the word
        frequencies.append(int(freq))  # Convert frequency to integer and append

# Set the figure size (width, height in inches)
plt.figure(figsize=(12, 6))  # Adjust size as needed, e.g., 12 inches wide and 6 inches tall

# Plotting the top 50 words
plt.bar(words, frequencies)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 50 Most Frequent Words')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust layout to prevent label cutoff

# Show the plot
plt.show()
```
